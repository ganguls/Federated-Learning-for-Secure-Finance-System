```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from datetime import datetime
import joblib
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 20)
```

```python
# Load Lending Club CSV
df = pd.read_csv("D:\FinalYrProject\FL System\original_dataset\loan.csv", low_memory=False)
# Quick overview
df.head()
df.info()
df.describe()
```

Drop Irrelevant Columns

```python
irrelevant_cols = ['id','member_id','url','desc','title']  # Add more if needed
df = df.drop(columns=irrelevant_cols)
```

```python
# Numeric columns  fill with median
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# Object/categorical columns  fill with "Unknown"
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].fillna("Unknown")
```

```python
# Ordinal encoding example
ordinal_cols = ['emp_length', 'sub_grade']
for col in ordinal_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    joblib.dump(le, f"{col}_encoder.pkl")  # Save encoder for reuse

# One-hot encoding for nominal columns
nominal_cols = ['grade','home_ownership','purpose','addr_state']
df = pd.get_dummies(df, columns=nominal_cols, drop_first=True)
```
